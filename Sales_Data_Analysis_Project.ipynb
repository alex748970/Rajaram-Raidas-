```python
ax = pd.merge(Detail , Order).groupby('State')['Profit'].sum()
print(ax)
```

```python
ax = pd.merge(Detail, Order).groupby('State')['Profit'].sum().nlargest(5).sort_values(ascending = True).plot(kind = 'bar')
sns.set_style('darkgrid')
plt.xlabel('State')
plt.ylabel('Profit')
plt.title('Sum of State By Profit')
for bar in ax.containers:
  ax.bar_label(bar)
plt.legend()
plt.show()
```

```python
merged_df = pd.merge(Detail, Order)
merged_df['Order Date'] = pd.to_datetime(merged_df['Order Date'], format='%d-%m-%Y')

daily_profit = merged_df.groupby('Order Date')['Profit'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(x='Order Date', y='Profit', data=daily_profit)
plt.title('Daily Profit Trend')
plt.xlabel('Order Date')
plt.ylabel('Total Profit')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
ax_profit = pd.merge(Detail, Order).groupby('State')['Profit'].sum().nlargest(5).sort_values(ascending = True).plot(kind = 'bar', color='skyblue')
sns.set_style('darkgrid')
plt.xlabel('State')
plt.ylabel('Profit')
plt.title('Top 5 States by Profit')
for bar in ax_profit.containers:
  ax_profit.bar_label(bar)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
ax_amount = pd.merge(Detail, Order).groupby('State')['Amount'].sum().nlargest(5).sort_values(ascending = True).plot(kind = 'bar', color='lightcoral')
sns.set_style('darkgrid')
plt.xlabel('State')
plt.ylabel('Amount')
plt.title('Top 5 States by Amount')
for bar in ax_amount.containers:
  ax_amount.bar_label(bar)
plt.tight_layout()
plt.show()
```
