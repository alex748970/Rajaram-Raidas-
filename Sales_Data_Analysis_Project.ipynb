```python
merged_df = pd.merge(Detail, Order)
merged_df['Order Date'] = pd.to_datetime(merged_df['Order Date'], format='%d-%m-%Y')

daily_profit = merged_df.groupby('Order Date')['Profit'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(x='Order Date', y='Profit', data=daily_profit)
plt.title('Daily Profit Trend')
plt.xlabel('Order Date')
plt.ylabel('Total Profit')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
merged_df = pd.merge(Detail, Order)
merged_df['Order Date']=pd.to_datetime(merged_df['Order Date'],format='%d-%m-%Y')
merged_df['Month_num'] = merged_df['Order Date'].dt.month
merged_df['Month'] = merged_df['Order Date'].dt.month_name()
monthly_profit = merged_df.groupby(['Month_num','Month'])['Profit'].sum().sort_index().reset_index(level='Month')

# Define colors based on profit values
colors = ['green' if profit > 0 else 'red' for profit in monthly_profit['Profit']]

# Plotting with conditional colors
ax = monthly_profit.plot(kind = 'bar', x='Month', y='Profit', figsize=(10, 6), color=colors)
plt.title('Monthly Profit Trend')
plt.xlabel('Month')
plt.ylabel('Profit')
plt.xticks(rotation=45)
plt.tight_layout()

# Add labels to the bars
for bar in ax.containers:
  ax.bar_label(bar)

plt.show()
```

```python
sns.set_style('darkgrid')

# Get the data for top 5 states by PaymentMode count
top_5_states_data = pd.merge(Detail, Order).groupby('State')['PaymentMode'].count().nlargest(5).sort_values(ascending=False)
top_5_states_df = top_5_states_data.reset_index()

plt.figure(figsize=(15, 6))

# Subplot 1: Bar Chart
plt.subplot(1, 2, 1) # 1 row, 2 columns, first plot
ax_bar = sns.barplot(x='State', y='PaymentMode', data=top_5_states_df, color='skyblue') # Changed palette to color
for bar in ax_bar.containers:
  ax_bar.bar_label(bar)
plt.xlabel('State')
plt.ylabel('Count PaymentMode')
plt.title('Top 5 States By Count PaymentMode (Bar Chart)')
plt.grid(False)

plt.subplot(1, 2, 2)
ax_line = sns.lineplot(x='State', y='PaymentMode', data=top_5_states_df, marker='o', color='coral')

for x, y in zip(top_5_states_df['State'], top_5_states_df['PaymentMode']):
    ax_line.text(x, y, y, color='black', ha='left', va='bottom')

plt.xlabel('State')
plt.ylabel('Count PaymentMode')
plt.title('Top 5 States By Count PaymentMode (Line Plot)')
plt.grid(True)

plt.tight_layout()
plt.show()
plt.savefig('Top 5 States By Count PaymentMode_plots')
```
